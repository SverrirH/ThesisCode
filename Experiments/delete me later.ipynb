```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
df_target = pd.read_csv('../../Data/Drainage data/Bo√∞agrandi_allt.csv')
df_target = df_target[df_target.loc[:,'CType'] == 4]
df_target.CTime = pd.to_datetime(df_target.CTime)
df_target = df_target[df_target.Val != 0]
```

```python
y_bod = df_target.loc[:,'Val']
y_bod.index = df_target.CTime
```

```python
plt.figure(figsize=(35,10))
plt.plot(y_bod,alpha = 0.5)
plt.plot(y_bod.rolling(168).mean())
plt.plot(y_bod.rolling(168).mean() - y_bod.rolling(168).std(),color='red',linestyle = '--')
plt.plot(y_bod.rolling(168).mean() + y_bod.rolling(168).std(),color='red',linestyle = '--')
```

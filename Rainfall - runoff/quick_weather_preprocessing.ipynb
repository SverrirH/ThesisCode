```python
df = pd.read_csv(URL_weather_data + 'Weather data 2020/from 2016 sept observations.csv')
df.head(3)
```

```python
df.station_id.value_counts()
```

```python
# Check how much NAN data is in each column for each station (checking which stations have certain observations)
name_dict = {
    1481 : 'Hólmsheiði',
    1475 : 'Reykjavík',
    7475 : 'Reykjavík búveðurstöð',
    1 : 'Reykjavík',
    1473 : 'Straumsvík',
    1479 : 'Korpa',
    7476 : 'Reykjavík sólgeislun',
    1477 : 'Reykjavíkurflugvöllur',
    6208 : 'Þykkvibær',
}
df.groupby('station_id').apply(lambda x : (~x.isna()).sum()/len(x)).T.round(2).rename(lambda x:name_dict[x] + ' ' + str(x),axis='columns',inplace=False)
# df.groupby('station_id').apply(lambda x : (~x.isna()).sum()/len(x)).T.round(2)
```

```python
df_rvk = df.loc[df.station_id.isin([1475,7476])]
```

```python
temperature = pd.pivot_table(df_rvk,values='temperature',index='date_time',columns='station_id').iloc[:,0]
temperature_min = pd.pivot_table(df_rvk,values='temperature_min',index='date_time',columns='station_id').iloc[:,0]
temperature_max = pd.pivot_table(df_rvk,values='temperature_max',index='date_time',columns='station_id').iloc[:,0]
radiation = pd.pivot_table(df_rvk,values='total_sun_radiation',index='date_time',columns='station_id').iloc[:,0]
moist_pressure = pd.pivot_table(df_rvk,values='moist_pressure',index='date_time',columns='station_id').iloc[:,0]
```

```python
column_names = ['temperature','temperature_min','temperature_max','radiation','moist_pressure']
df_input = pd.DataFrame([temperature,temperature_min,temperature_max,radiation,moist_pressure],index=column_names).T
```

```python
df_input.index = pd.DatetimeIndex(df_input.index)
df_input = df_input.resample('1h').interpolate(method='linear')
df_input.to_pickle('weather_data')
```
